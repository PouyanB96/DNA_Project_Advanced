```python
import pandas as pd
from Bio import SeqIO
import matplotlib.pyplot as plt
DNA_data = []
for i in SeqIO.parse("example.fasta", "fasta"):
    #print(i.id)
    #print(i.seq)
    seq = str (i.seq).upper()
    #print (seq)
    A = seq.count ('A')
    T = seq.count ('T')
    C = seq.count ('C')
    G = seq.count ('G')
    CG = round(((C + G) / (C + G + T + A)) *100  , 2)
    DNA_data.append({
        "Sequence ID": i.id,
        "A": A,
        "T": T,
        "G": G,
        "C": C,
        "CG%": CG
    })
DNA_chart = pd.DataFrame(DNA_data)
print (DNA_chart)
DNA_chart.plot(
    x="Sequence ID",
    kind="bar"
)
plt.title("GC Content per Sequence")
plt.ylabel("Sequence Features")
plt.xlabel("Sequence ID")
plt.tight_layout()
plt.show()
```
